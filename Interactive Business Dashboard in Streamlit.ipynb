**Import Libraries**

```python
import streamlit as st
import pandas as pd
import plotly.express as px
```

**Load Dataset**

```python
df = pd.read_csv("Global_Superstore2.csv", encoding='latin1')
```

**Clean Dataset**

```python
# Remove missing values
df.dropna(inplace=True)

# Convert order date
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

**Create Dashboard Title**

```python
st.title("Global Superstore Dashboard")
```

**Create Sidebar Filters**

Region Filter

```python
 region = st.sidebar.multiselect(
    "Select Region",
    options=df['Region'].unique(),
    default=df['Region'].unique()
)
```

Category Filter

```python
category = st.sidebar.multiselect(
    "Select Category",
    options=df['Category'].unique(),
    default=df['Category'].unique()
)
```

Sub-Category Filter

```python
sub_category = st.sidebar.multiselect(
    "Select Sub-Category",
    options=df['Sub-Category'].unique(),
    default=df['Sub-Category'].unique()
)
```

**Filter Dataset**

```python
filtered_df = df[
    (df['Region'].isin(region)) &
    (df['Category'].isin(category)) &
    (df['Sub-Category'].isin(sub_category))
]
```

**Display KPIs**

Total Sales

```python
total_sales = filtered_df['Sales'].sum()

st.metric("Total Sales", f"${total_sales:,.2f}")
```

Total Profit

```python
total_profit = filtered_df['Profit'].sum()

st.metric("Total Profit", f"${total_profit:,.2f}")
```

**Top 5 Customers by Sales**

```python
top_customers = filtered_df.groupby('Customer Name')['Sales'].sum().nlargest(5).reset_index()

fig_customer = px.bar(
    top_customers,
    x='Customer Name',
    y='Sales',
    title='Top 5 Customers by Sales'
)

st.plotly_chart(fig_customer)
```

**Plot Top Customers Chart**

```python
fig_customer = px.bar(
    top_customers,
    x='Customer Name',
    y='Sales',
    title='Top 5 Customers by Sales'
)

st.plotly_chart(fig_customer)
```

**Sales by Region**

```python
region_sales = (
    filtered_df.groupby('Region')['Sales']
    .sum()
    .reset_index()
)
```

Plot Chart

```python
fig_region = px.pie(
    region_sales,
    names='Region',
    values='Sales',
    title='Sales by Region'
)

st.plotly_chart(fig_region)
```

**Profit by Category**

```python
category_profit = (
    filtered_df.groupby('Category')['Profit']
    .sum()
    .reset_index()
)
```

**Plot Chart**

```python
fig_profit = px.bar(
    category_profit,
    x='Category',
    y='Profit',
    title='Profit by Category'
)

st.plotly_chart(fig_profit)
```
